## Import Data

```python
import pandas as pd
import numpy as np
```

```python
grid = pd.read_csv("data/area/CSV_file/sqaure_grid.csv")
grid = grid.set_index("id")
grid
```

### Designated Green Space

```python
dgs = pd.read_csv("data/area/CSV_file/designated_green_space.csv")
dgs
```

```python
#Calculate the percentage of green space + Clean the table

dgs["dgs%"]=dgs.apply(lambda x:x["gp_area(m2)"]/x["area(m2)"]*100, axis=1).fillna(0)
dgs = dgs.drop(columns=["left", "top", "right", "bottom", "area(m2)","%_area"]).rename(columns={"gp_area(m2)":"dgs_area"})
dgs=dgs.set_index("id")
dgs
```

### Quasi Public Private Green Space

```python
qgs = pd.read_csv("data/area/CSV_file/quasi_green_space.csv")
qgs.head(5)
```

```python
qgs["qgs%"]=qgs.apply(lambda x:x["q_gp_area(m2)"]/x["grid_area"]*100, axis=1).fillna(0)
qgs = qgs.drop(columns=["left", "top", "right", "bottom", "grid_area", "id_2"])
qgs = qgs.rename(columns={"q_gp_area(m2)":"qusai_area"})
qgs = qgs.set_index("id")
qgs
```

### Water Green Space

```python
water = pd.read_csv("data/area/CSV_file/waterbodies_green_space.csv")
water.head(5)
```

```python
water["water%"]=water.apply(lambda x:x["water_area"]/x["grid_area"]*100, axis=1).fillna(0)
water=water.drop(columns=["left", "top", "right", "bottom", "grid_area","id_2"])
water=water.set_index("id")
water
```

### Natural Wild Green Space

```python
wild = pd.read_csv("data/area/CSV_file/wild_green_space.csv")
wild.head(5)
```

```python
wild["wild%"]=wild["wn_area(m2)"]/wild["grid_area"]
wild=wild.drop(columns=["left", "top", "right", "bottom", "grid_area", "id_2"])
wild = wild.rename(columns={"wn_area(m2)":"wild_area"})
wild = wild.set_index("id")
wild
```

### Total Green Space

```python
total = pd.read_csv("data/area/CSV_file/total_green_space.csv")
total.head()
```

```python
total["tgp_area%"]=total.apply(lambda x:x["total_gp_area"]/x["grid_area"]*100, axis=1).fillna(0)
total = total.drop(columns=["left", "top", "right", "bottom", "grid_area"])
total = total.set_index("id")
total
```

## Joining the Green Space DFs to the Main Grid DF

```python
grid
```

```python
join1 = grid.join(total, how="left")
join2 = join1.join(dgs, how="left")
join3=join2.join(qgs, how="left")
join4=join3.join(water, how="left")
final=join4.join(wild, how="left")
```

```python
final = final.fillna(0)
final.to_csv("data/area/CSV_file/greenspace_final.csv")
```
