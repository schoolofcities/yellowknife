```python
import os
import csv
import pandas as pd
import geopandas as gpd

df = pd.read_csv("CO2e Calculations.csv")
add = gpd.read_file("yellowknife-da-boundary.geojson")
```

```python
add_sub = add[["dauid" , "landarea" , "geometry"]]
add_sub.head()
```

```python
df.head()
```

```python
df['dauid'] = df['dauid'].astype(int)
add_sub['dauid'] = add_sub['dauid'].astype(int)
```

```python
mer = add_sub.merge(df , left_on = 'dauid' , right_on = 'dauid')
mer
```

```python
mer.plot()
```

```python
mer.to_file("nwt_final.geojson" , driver = 'GeoJSON')
mer.to_csv('test.csv' , index = False)
```

```python
mer.to_file('nwt_final.shp' , driver = 'ESRI Shapefile')
```
